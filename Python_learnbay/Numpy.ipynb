# concatination of array/ stacking of array---> joining of an array

```python
a=np.array([['new york','maryland'],['Albany','Annapolis']])
b=np.array([['ohio','Arizona'],['Columbus','Phoenix']])
```

```python
np.concatenate((a,b)) # (a,b)--tuple of array
```

```python
"""
axis=0  'row wise' --> (no of columns for both matrix should be same)
axis=1  'Column wise'--> (no of rows should be same for both the matrix)


"""


np.concatenate((a,b),axis=1)
```
