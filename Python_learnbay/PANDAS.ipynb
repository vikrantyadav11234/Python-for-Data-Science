# merge: add the data based on some common column

* 1st rule: can only merge two dataframes at a time.
* 2nd: atleast one column should be common between the two dataframe.

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1,df2,on="city")
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","Baltimore"],
    "temperature": [21,14,35,38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","sen diago"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1,df2,on="city")
```

```python
# left join

pd.merge(df1,df2,on="city", how="left") # common plus left table(first table)
```

```python
# right join
pd.merge(df1,df2,on="city", how="right") #common data plus all the left in table 2
```

```python
# outer join
pd.merge(df1,df2,on="city", how="outer") # common data plus table 1 plus table 2 remaining
```

```python
pd.merge(df1,df2,on="city", how="outer", indicator=True)
```
