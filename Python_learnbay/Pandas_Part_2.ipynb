# normal slicing
- column slicing is not possible
- row slicing is possible
- zero indexing is possible

# loc
- both row and column slicing is possible
- drawback - it is limited only to the row and column name
- zero indexing is not possible

# iloc
- both row and column slicing is possible
- zero indexing is possible
- custom names are not allowed for slicing

# Iterating over a Dataframe

1. itertuple
2. iterrows
3. iteritems

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
df
```

# 1. itertuples

```python
for i in df.itertuples():
    print(i)
```

```python
df
```

```python
for i in df.itertuples():
    if i.Organizations_name == 'CNES':
        print(i.Headquarters)
```

# 2. iterrows

```python
df
```

```python
for i in df.iterrows():
    print(i)
```

```python
(0 , Organizations_name       CNSA
Headquarters          Beijing
Country_name            China
Formation_year           1993
Name: 4, dtype: object)
```

```python
tup = [(0 , 'python') , (1, 'learnbay')]

for i in tup:
    print(i[1])
```

```python
for i in df.iterrows():
    print(i[1], end = '\n\n\n')
```

```python
for i in df.iterrows():
    print(i[1]['Organizations_name'])
```

```python
for i in df.iterrows():
    if i[1]['Organizations_name'] == 'ISRO':
        print(i[1]['Headquarters'])
```

```python
for i in df.itertuples():
    if i.Organizations_name == 'ISRO':
        print(i.Headquarters)
```

# Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df = pd.concat([df1 , df2])
df
```

```python
df = pd.merge(df1 , df2 , how = 'inner')
df
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
# inner join
df = pd.merge(df1 , df2 , how = 'inner')
df
```

```python
# outer join
df = pd.merge(df1 , df2 , how = 'outer')
df
```

```python
# left join
df = pd.merge(df1 , df2 , how = 'left')
df
```

```python
df1
```

```python
df2
```

```python
# right join
df = pd.merge(df1 , df2 , how = 'right')
df
```

```python
# outer join
df = pd.merge(df1 , df2 , how = 'outer' , indicator= True)
df
```
