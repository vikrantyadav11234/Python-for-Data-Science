# fill the null values
* fill age with median
* fill Cabin with mode
* fill Embarked with mode

```python
data.fillna({
    "Age":data["Age"].median(),
    "Cabin":data["Cabin"].mode()[0],
    "Embarked":data["Embarked"].mode()[0]
}, inplace=True)
```

```python
sex_list=list(data['Sex'])
male_sex_count=0
female_sex_count=0
sex_list_label=['male/female']
for i in sex_list:
    if i=='male':
        male_sex_count=male_sex_count+1
    else:
        female_sex_count=female_sex_count+1
male_sex_count=[male_sex_count]
female_sex_count=[female_sex_count]
plt.bar(sex_list_label, male_sex_count, color='r')
plt.bar(sex_list_label, female_sex_count, color='b')
# plt.show()
```

```python
# if the age is >=18 adult
# if the age is <18 child
```

```python
emp_list=[]
for i in data.iterrows():
    if i[1]['Age']>=18:
        emp_list.append('Adult')
    else:
        emp_list.append('Child')
```

```python
emp_list
```

```python
data['Adult']=emp_list
```
