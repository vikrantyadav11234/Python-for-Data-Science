```python
class my_exception(Exception):
    def __init__(self , msg):
        self.msg = msg
        
def submit_pan(pan):
    if len(pan) == 10:
        print('we will verify it and deposit')
    else:
        raise my_exception('Wrong Pan number')
        
def Amount_deposited(amt):
    if amt > 50000:
        raise my_exception('Over limit')
    else:
        print('Amount deposited')
```

```python
amt = int(input('Please enter the amount'))

try:
    Amount_deposited(amt)
except my_exception as e:
    print(e)
    try:
        pan = input('Enter the pan number')
        submit_pan(pan)
    except my_exception as e:
            print(e)
```
