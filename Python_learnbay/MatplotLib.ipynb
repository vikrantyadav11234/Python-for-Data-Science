# matplotlib- library used for data visualization

* To get batter understanding of the data


```python
import matplotlib.pyplot as plt
```

```python
week_day=[1,2,3,4,5,6,7]
temp=[23,34,24,56,21,38,45]
```

```python
# which day has the minimum temp
# in matplotlib we can plot 1D and 2D data only
# when we want to plot 2D data it will take 2 values one is x and other is y
plt.plot(week_day,temp)
```

```python
plt.plot(week_day,temp, color='g')
plt.ylabel('temp')
plt.xlabel('week_day')
plt.title('temp vs week_day')
```

# Bar graph

* one data should be catagorical and one data should be continuous.
  ex of catagorical data: good, bad, excellent etc.


```python
company=['Amazon','Flipkart','Google','microsoft']
revenue=[24,50,150,120]
```

```python
# vertical plot
plt.bar(company,revenue)
```

```python
plt.bar(company,revenue,color='g')
plt.ylabel('company')
plt.xlabel('revenue')
plt.title('Revenue vs Company')
```

```python
# horizontal plot
plt.bar(company,revenue,color='g')
plt.ylabel('company')
plt.xlabel('revenue')
plt.title('Revenue vs Company')
```

```python
blood_sugar=[113,85,90,95,102,123,134,152,77,52,63,96,86,59,83]
plt.hist(blood_sugar)

```

```python
# pie chart
exp_labels=['home rent','food','internet bill','car','other utilities']
exp_values=[1400,600,400,410,250]
```

```python
plt.pie(exp_values, labels= exp_labels)
```

```python
exp_labels=['home rent','food','internet bill','car','other utilities']
exp_values=[1400,600,400,410,250]
plt.pie(exp_values, labels= exp_labels,autopct='%0.2f%%')
```

```python
exp_labels=['home rent','food','internet bill','car','other utilities']
exp_values=[1400,600,400,410,250]
plt.pie(exp_values, labels= exp_labels,autopct='%0.2f%%', explode=[0.1,0,0,0,0.08])
```

# Scatter plot

```python
plt.scatter(x,y)
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[4,5,7,3,7,8,2,4,6]
plt.scatter(x,y)
plt.grid()
```
