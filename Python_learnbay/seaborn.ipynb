```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

# Seaborn link: https://github.com/mwaskom/seaborn-data

```python
# relational plot- x & y
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips , hue='smoker') 
# hue always takes a catagorical column and
# hue always catagorise everything based on color
```

```python
sns.relplot(x='total_bill',y='tip',data=tips , hue='smoker',style='time')
# style always takes a catagorical column and
# style will tell in different different style-ex- square ,cross, or anything
```

```python
sns.relplot(x='total_bill',y='tip',data=tips , hue='smoker',style='time',size='sex')
#size(catagorical)(size)
```
