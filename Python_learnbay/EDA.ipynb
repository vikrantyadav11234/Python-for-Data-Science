# How to treat the outliers
- 1- As per the quantile method we cap the values between 1st and 99th 
     percentile values.
- 2- As per the IQR method we cap the values between upper bound and lower
 bound values.
## Caping
- 1- all values lower than the lower bound value will be kept at lower bound
     value and all values upper than the upper bound value will be kept at 
    upper bound value.

```python
# We are using IQR to cap the value in age column
q1=data.age.quantile(0.25)
#q2=data.age.quantile(0.50)
q3=data.age.quantile(0.75)
iqr=q3-q1

print(q1,q3,iqr, sep='\n')
lower=q1-1.5*iqr
upper=q3+1.5*iqr

print(lower,upper, sep='\n')
```

# to cap the values we can use np.where() method
- 1- np.where()
# clip function
- 2- clip()

```python
data.age=np.where(data.age<lower,lower, np.where(data.age>upper,upper,data.age))
```

### Treatment of the outliers in the num data

```python
cols=list(num_data.columns)
```

```python
cols
```

```python
def out_caping(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1

    
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr

    return data.clip(lower=lower,upper=upper)
```

```python
for i in cols:
    if i!='churn':
        data[i]=out_caping(data[i])
```
