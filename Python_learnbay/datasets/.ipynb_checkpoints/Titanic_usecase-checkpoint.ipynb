```python
# IQR
# Z-Score
```

```python
# Z- score

mean_of_fare = np.mean(data['fare'])
std_of_fare = np.std(data['fare'])

threshold = 3
outlier = []

for i in data['fare']:
    z = (i - mean_of_fare)/(std_of_fare)
    if z > threshold:
        outlier.append(i)

print('Total no of outliers is : ' , len(outlier))
print('Minimu value of outliers is : ', min(outlier))
print('max value of outliers is : ', max(outlier))
```

```python
len(outlier) / len(data['fare']) * 100
```

```python
sns.displot(x = outlier ,kde = True).set(title = 'Outlier distribution')
```

```python
# IQR


Q1,Q3 = data.fare.quantile([0.25 , 0.75])
print(Q1)
print(Q3)

IQR = Q3 - Q1

print(IQR)


lower_outlier = Q1 - 1.5*(IQR)
upper_outlier = Q3 + 1.5*(IQR)

print(lower_outlier)
print(upper_outlier)
```

# Drop the outliers

```python
to_drop = data.loc[data['fare'] > min(outlier)]
to_drop
```

```python
data.drop(to_drop.index , inplace = True)
```
