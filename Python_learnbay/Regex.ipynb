```python
# it is a module which we need to import it 
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```

```python
import re
```

```python
#step 1
pattern=re.compile(r'M')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 2
pattern=re.compile(r'Mr')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 3
pattern=re.compile(r'Mr\.')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 4
pattern=re.compile(r'Mr\.\s')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 5
pattern=re.compile(r'Mr\.\s[A-Z]')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 6 # quantifiers
pattern=re.compile(r'Mr\.\s[A-Z][a-z]*')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 7 # quantifiers
pattern=re.compile(r'Mr\.?\s[A-Z][a-z]*')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# step 8 # quantifiers
pattern=re.compile(r'M[rs]\.?\s[A-Z][a-z]*')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# either and group works
# step 7 # quantifiers
pattern=re.compile(r'M(r|s|rs)\.?\s[A-Z][a-z]*')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# extract a proper mail
emails='''
Afsanakhan@gmail.com
vikranty622.122@university.edu
vikrant.ecd.cf@ietlucknow.ac.in
afsana-321-khan@my-work.net
afsana-345-khan@my-work.abc
'''

# pattern=re.compile(r'M(r|s|rs)\.?\s[A-Z][a-z]*')
pattern=re.compile(r'[A-za-z]*(0-9|.|@)*[^-]')
matches=pattern.finditer(emails)
for match in matches:
    print(match)
```
